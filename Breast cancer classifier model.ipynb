# Importing Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('breast cancer data.csv')
```

```python
df.drop(['id', 'Unnamed: 32'], axis =1, inplace=True)
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
```

# Preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay, precision_score, recall_score, RocCurveDisplay
```

```python
x=df.drop('diagnosis', axis=1)
```

```python
y=df['diagnosis']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
```

#### Logistic Regression

```python
lr_model = LogisticRegression()
```

```python
lr_model.fit(x_train_scaled, y_train)
```

```python
lr_pred = lr_model.predict(x_test_scaled)
```

```python
accuracy_score(y_test, lr_pred)
```

```python
print(classification_report(y_test, lr_pred))
```

### SVM

```python
svc_model = SVC()
```

```python
param_grid={
    'C':[0.01, 0.1, 1],
    'kernel':['linear', 'rbf', 'poly'],
    'degree':[2, 3],
    'gamma':['scale', 'auto']
}
```

```python
svc_grid_model = GridSearchCV(svc_model, param_grid=param_grid)
```

```python
svc_grid_model.fit(x_train_scaled, y_train)
```

```python
svc_pred = svc_grid_model.predict(x_test_scaled)
```

```python
svc_grid_model.best_params_
```

```python
svc_grid_model.best_score_
```

```python
accuracy_score(y_test, svc_pred)
```

```python
print(classification_report(y_test, svc_pred))
```

### Random Forest

```python
rf_model = RandomForestClassifier(random_state=1)
```

```python
param_grid={
    'n_estimators':[32, 64, 100, 128, 150],
    'max_features':[4, 8, 16],
    'max_depth':[4, 8, 16],
    'bootstrap':[True, False],
}
```

```python
rf_grid_model = GridSearchCV(rf_model, param_grid=param_grid)
```

```python
rf_grid_model.fit(x_train, y_train)
```

```python
rf_pred = rf_grid_model.predict(x_test)
```

```python
rf_grid_model.best_params_
```

```python
rf_grid_model.best_score_
```

```python
accuracy_score(y_test, rf_pred)
```

```python
print(classification_report(y_test, rf_pred))
```

### Gradient Boosting

```python
gb_model = GradientBoostingClassifier()
```

```python
param_grid={
    'learning_rate':[0.01, 0.05, 0.1, 0.5],
    'n_estimators':[32, 64, 100, 128, 150],
    'max_features':[4, 8, 16],
    'max_depth':[4, 8, 16],
}
```

```python
gb_grid_model = GridSearchCV(gb_model, param_grid=param_grid)
```

```python
gb_grid_model.fit(x_train, y_train)
```

```python
gb_pred = gb_grid_model.predict(x_test)
```

```python
gb_grid_model.best_params_
```

```python
gb_grid_model.best_score_
```

```python
accuracy_score(y_test, gb_pred)
```

```python
print(classification_report(y_test, gb_pred))
```

```python
acc_score=[]
prec_score=[]
rec_score=[]
def scores(y_pred):
    acc_score.append(round(accuracy_score(y_test, y_pred), 2))
    prec_score.append(round(precision_score(y_test, y_pred), 2))
    rec_score.append(round(recall_score(y_test, y_pred), 2))

scores(lr_pred)
scores(svc_pred)
scores(rf_pred)
scores(gb_pred)
```

```python
scores=[acc_score, prec_score, rec_score]
```

```python
np.array(scores).T
```

```python
pd.DataFrame(data=np.array(scores).T, columns=['accuracy', 'precision', 'recall'], index=['Logistic Reg', 'Svc', 'Random Forest', 'Gradient Boosting'])
```

```python

models = ["Logistic Reg", "SVC", "Random Forest", "Gradient Boosting"]

# Set width of bars
bar_width = 0.2
x = np.arange(len(models))  # X locations for groups

# Create bars
plt.figure(figsize=(10, 5))
plt.bar(x, acc_score, width=bar_width, label="Accuracy", color="skyblue")
plt.bar(x + bar_width, prec_score, width=bar_width, label="Precision", color="orange")
plt.bar(x + 2 * bar_width, rec_score, width=bar_width, label="Recall", color="green")

# Labels & Formatting
plt.xlabel("Models")
plt.ylabel("Score")
plt.title("Model Performance Comparison")
plt.xticks(x + bar_width, models)  # Adjust x-axis ticks to align with bars
plt.ylim(0.85, 1.05)  # Set y-axis limit
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot
plt.show()
```

```python
fig, ax=plt.subplots()
RocCurveDisplay.from_estimator(lr_model, x_test_scaled, y_test, ax=ax, name='Logistic Reg')
RocCurveDisplay.from_estimator(svc_grid_model, x_test_scaled, y_test, ax=ax, name='SVC')
RocCurveDisplay.from_estimator(rf_grid_model, x_test, y_test, ax=ax, name='Random Forest')
RocCurveDisplay.from_estimator(gb_grid_model, x_test_scaled, y_test, ax=ax, name='Gradient Boosting')
```
